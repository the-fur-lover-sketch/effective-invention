```python
#
n=1
#var_dict={
#  "t" : [1],
#  "sp": [1],
#  "f" : [1],
#  "dr": [1],
#  "da": [1],
# "wave": [1]}

t=1
n=1
sp=1
f=1
dr=1
da=1
wave=1

var_dif_char="""
#---variables---#

t_1=3
t_2 = 0.125
sp_1=0
f_1=110
dr_1=0.6
da_1=0
wave_1='sine'

#---score data---#

"""


sheet_music=""
sheet_music_lists=[[var_dif_char]]
next_lines = ""
orders=[]
order = ""
notes_list =[]
stop = False



while True:
  if stop == True:
    break
  user_input=input("input orders:")

  #入力を"_"で分ける
  orders = user_input.split("_")


  #命令ごとに処理
  for order in orders:

    # ①停止処理
    if order == "stop":
        print("func_stop executed")
        stop = True
        break

    # ④出力命令
    if order == "print":

      result = "\n".join(row[0] for row in sheet_music_lists)
      print("here's ur sheet-music : \n",result)
      continue

    # ⑤変数情報出力
    if order == "ls":
      print(f"func_ls executed \n t_{t} \n n:{n} \n sp_{sp} \n f_{f} \n dr_{dr} \n da_{da} \n wave:{wave}")
      continue

    # ⑥やり直し
    if order == "reverse":
      print("\nfunc_reverse executed\n")
      sheet_music_lists = sheet_music_lists[:-1]
      continue

    # ②周波数入力の場合("="がない場合)
    if order.count("=") == 0:

      #入力エラー
      try:
        notes_list=[int(note) for note in order.split()]
      except:
        print("failed to read input")
        break
      if order == "":
        continue

      #周波数をリストにして代入
      print("notes_list",notes_list)
      next_lines = next_lines + f"\nt_{t}*{n} + sp_{sp} , f_{f}*{notes_list} , dr_{dr}*t_{t} + da_{da} , '{wave}'"
      n += 1


    # ③変数代入の場合("="がひとつ)
    if order.count("=") == 1:
      print("split order:",order.split("="))

      #入力エラー
      try:
        change_var , val = order.split("=")
      except:
        print("failed to read input")

      #変数が存在するかチェックしてから代入
      if change_var in globals():
        try:
          globals()[change_var] = int(val)
          print(f"{change_var} was changed to: {val}")
        except:
          print(f"could not change {change_var}")
      else:
        print(f"{change_var} was not found")
      continue

  if next_lines != "":
    print("added next_lines:" , next_lines)
    sheet_music_lists.append([next_lines[1:]])
    next_lines=""
```
